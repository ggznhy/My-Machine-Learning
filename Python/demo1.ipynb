```python
# "append() adds an object as a single element, while extend() adds each element of an iterable individually."

lst = [1, 2, 3]
lst.append([4, 5])
print(lst)  # Output: [1, 2, 3, [4, 5]]

lst = [1, 2, 3]
lst.extend([4, 5])
print(lst)  # Output: [1, 2, 3, 4, 5]
```
