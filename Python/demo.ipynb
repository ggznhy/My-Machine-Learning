```python
# Assignment 8.5
# Open the file
file = open('mbox-short.txt')

# Initialize a counter
count = 0

# Read the file line by line
for line in file:
    # Skip lines that don't start with 'From '
    if not line.startswith('From '):
        continue

    # Split the line into words
    words = line.split()

    # Print the email address (second word)
    print(words[1])

    # Increment the counter
    count += 1

# Print the count at the end
print(f"There were {count} lines in the file with From as the first word")
```

```python
# Assignment 9.4

name = input("Enter file:")
if len(name) < 1:
    name = "mbox-short.txt"
handle = open(name)

box = {}

for line in handle:
    if not line.startswith('From '):
        continue

    eml = line.split()[1]

    # print(eml)

    box[eml] = box.get(eml, 0) + 1


km = ''
max = -1

for k in box:
    # print(k)
    if box[k] > max:
        km = k
        max = box[k]

print(km, max)
```

```python
# Assignment 10.2
name = input("Enter file:")
if len(name) < 1:
    name = "mbox-short.txt"
handle = open(name)

hour_counts = {}

for line in handle:
    if not line.startswith('From '):
        continue

    hour = line.split()[5][0:2]
    hour_counts[hour] = hour_counts.get(hour, 0) + 1

# Sort the hours and print the counts
for hour in sorted(hour_counts):
    print(hour, hour_counts[hour])
```
